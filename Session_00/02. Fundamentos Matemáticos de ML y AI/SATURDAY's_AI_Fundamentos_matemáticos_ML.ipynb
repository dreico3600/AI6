**Una ecuación matricial** es una ecuación de la forma:
$Ax=b$
donde $A$ es una matriz con coeficientes en los n\'umeros reales.

El número de renglones de $A$ se denota por $n$ y el número de columnas se denota por $m$. 

$x$ denota el vector de incógnitas y $b$ es un vector. 

El vector $x$ tiene $m$ renglones y 1 columna.

El vector $b$ tiene $n$ renglones y 1 columna.




El problema de la producción de pokemones

$$3x+y=10,000\\ 4x+3y=20,000$$


Ecuación  matricial

$$\begin{pmatrix}
3 &1\\
4 &3
\end{pmatrix} 
\begin{pmatrix}
x_1\\x_2
\end{pmatrix}=\begin{pmatrix}
10000\\
20000
\end{pmatrix}$$









Solución:


A.  Forma Estándar:
Despejas, sustituyes y resuelves....


1. Despejar de la primera ecuación:

$$y=10000-3x$$

2. Sustituir en la segunda 
$$4x+3(10000-3x)=20000$$


3. Despejas x
$$x=2000$$

4. Sustituyes en despeje de $y$
$$y=4000$$

5. Concluyes

La solución es:

$$\begin{pmatrix}
x \\
y
\end{pmatrix} 
=\begin{pmatrix}
2000\\
4000
\end{pmatrix}$$






B. Gauss-Jordan

1. Usas la matriz aumentada del sistema
2. Encuentras matrices equivalentes
3. Resuelves

$$\begin{pmatrix}
3 &1 &| &10000\\
4 &3 &| & 20000
\end{pmatrix}\sim \cdots \begin{pmatrix}
1 &0 &| &2000\\
0 &1 &| & 4000
\end{pmatrix}$$



La solución es:

$$\begin{pmatrix}
x \\
y
\end{pmatrix} 
=\begin{pmatrix}
2000\\
4000
\end{pmatrix}$$

Inverso multiplicativo de un número $n$: $1/n$, si $n\neq 0$

**Matriz inversa**

**Def** Se dice que una matriz $A$ de $n\times n$ es invertible si existe otra matriz $C$ de $n\times n$ tal que $$CA=I_n, AC=I_n$$


donde $I_n=[d_{ij}]$, $d_{ij}=1$ si $i=j$ y  $d_{ij}=0$ si $i\neq j
La matriz inversa de $A$ es única, así denotamos $C=A^{-1}$ 


Una matriz que no es invertible en ocasiones se llama matriz singular, y una matriz invertible
se llama matriz no singular.

** Matriz inversa y sistema de ecuaciones**

Consideremos el sistema 

\begin{eqnarray}
Lx&=&b \qquad\text{donde} \quad 
L=\begin{pmatrix}
l_{11} &0  & \cdots\\
l_{21} & l_{22} & \\
\vdots & \vdots &\vdots\\
l_{n1} & \cdots & l_{nn} 
\end{pmatrix}
\end{eqnarray}


**Proposicion**

Sea $L$ una matriz triangular inferior. El sistema $Lx=b$ tiene solución si y sólo si los elementos de la diagonal de $L$ son diferentes de cero.

Notemos que las ecuaciones asociadas al sistema son

\begin{eqnarray}
l_{11}x_1&=&b_1\\
l_{21}x_1+l_{22}x_2&=&b_2\\
&\vdots&\\
l_{n1}x_1 +\cdots+ l_{nn} x_n &=&b_n
\end{eqnarray}

Asi, podemos calcular 

\begin{eqnarray}
x_1&=&\frac{b_1}{l_{11}}\\
x_2&=&\frac{1}{l_{22}}\left(b_2-l_{21}x_1\right)\\
&\vdots&\\
x_n &=&\frac{b_n}{l_{nn}}(l_{n1}x_1 +\cdots+ l_{n(n-1)})x_{n-1}
\end{eqnarray}

Por tanto, el sistema $Lx=b$ tiene solución si $L$ es una matriz 
invertible. Y la solución del sistema es $x=L^{-1}b$

El teorema de la matriz invertible
Sea A una matriz cuadrada de $n\times n$. Entonces, los siguientes enunciados son equivalentes. Es decir, para una A dada, los enunciados son todos ciertos o todos falsos.

a) A es una matriz invertible.

b) A es equivalente por filas a la matriz identidad de $n\times n$.

c) A tiene n posiciones pivote.

d) La ecuación $Ax= 0$ tiene solamente la solución trivial.

e) Las columnas de A forman un conjunto linealmente independiente.

f) La transformación lineal $x\rightarrow Ax$ es uno a uno.

g) La ecuación $Ax = b$ tiene al menos una solución para toda b en $\mathbf{R}^n$ .

h) Las columnas de A generan $\mathbf{R}^n$ .

i) La transformación lineal $x\rightarrow Ax$ mapea $\mathbf{R}$ sobre$\mathbf{R}$ .

j) Existe una matriz C de $n\times n$ tal que $CA =I$,

k) Existe una matriz D de $n\times n$ tal que $AD=I$.

l) $A^T$ es una matriz invertible.

Digamos que en tu ciudad 1 de cada 1000 personas tienen covid-19. Supongamos que una persona se siente maly se realiza el test.




Usando que el 1% de los resultados es falso positivo y 10%  falso negativo. 


¿Cuál piensas que es la probabilidad de que esa persona tenga covid-19?




confusionMatrxiUpdated.jpg


Source:https://manisha-sirsat.blogspot.com/2019/04/confusion-matrix.html

Cuál es la probabilidad de que un individuo que da positivo en el test de covid 19 tenga en realidad la enfermedad?



P(D): Probabilidad de tener la enfermedad (covid-19)


P(\neg D): Probabilidad de NO tener la enfermedad (covid-19)


P(+): Probabilidad de que el test sea positivo


*P(D|+): Probabilidad de tener la enfermedad dado que el test fue positivo  (TP)

P(\neg D|+): Probabilidad de NO tener la enfermedad dado que el test fue positivo  (FP)



1% or 0.01 FP (dado)
10% or 0.1 FN (dado)





Matriz de confusión

Teorema de Bayes


$$P(D|+)=\frac{P(+\cap D)}{P(+)}$$


Probabilidad Total

$$P(+)=P(+\cap D)+P(+\cap \neg D)$$




$P(D|+) = P(+|D) * P(D) / P(+|D) * P(D) + P(+|\neg D) * P(\neg D)$

P(D|+) = 0.9 * 0.001 / 0.9 * 0.001 + 0.01 * 0.999

P(D|+) = 0.0009 / 0.0009 + 0.00999

P(D|+) = 0.0009 / 0.01089

P(D|+) ~ 0.08264 or 8.26%

```python
from random import random, seed
seed(0)
pop = 10000000 # 10M people
counts = {}
for i in range(pop):
    has_covid = i % 1000 == 0 # one in 1,000 people have covid (priors or prevalence of disease)
    # The major assumption is that every person gets tested regardless of any symptoms
    if has_covid:                  # Has disease
        tests_positive = True      # True positive
        if random() < 0.1:     
            tests_positive = False # False negative
    else:                          # Does not have disease
        tests_positive = False     # True negative
        if random() < 0.01:    
            tests_positive = True  # False positive
    outcome = (has_covid, tests_positive)
    counts[outcome] = counts.get(outcome, 0) + 1
for (has_covid, tested_positive), n in counts.items():
    print('has covid: %6s, tests positive: %6s, count: %d' % (has_covid, tested_positive, n))
n_positive = counts[(True, True)] + counts[(False, True)]
print('number of people who tested positive:', n_positive)
print('probability that a test-positive person actually has covid: %.2f' % (100.0 * counts[(True, True)] / n_positive), )
```

Nested in the for-loop are if-statements that segment the population (10M) into one of four categories True Positive, False Negative, True Negative, False Positive.
