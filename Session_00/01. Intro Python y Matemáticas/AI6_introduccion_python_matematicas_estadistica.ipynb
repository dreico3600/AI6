## Orden de las operaciones aritm√©ticas

El orden en que se llevan a cabo las operaciones aritm√©ticas en `Python` es el siguiente:

- Primero se calcula lo que se halla entre par√©ntesis.
- A continuaci√≥n, las potencias.
- Despu√©s, productos y divisiones. En caso de haber varias, el orden que se sigue es de izquierda a derecha.
- Finalmente, sumas y restas. En caso de haber varias, el orden que se sigue es de izquierda a derecha.

```python
6 + 2 * 8 / 4 - 2 ** 3
```

```python
(6 + 2) * (8 / (4 - 2)) ** 3
```

```python
(6 + 2) * 8 / (4 - 2) ** 3
```

**Observaci√≥n.** El uso de los par√©ntesis puede cambiar completamente el resultado. No conviene abusar de ellos, aunque es mejor que sobren, ya que ayudan a entender el orden en que se van a llevar a cabo las operaciones.

# Estad√≠stica
Muestreo estad√≠stico



*   Poblaci√≥n : Todos los sujetos 
*   Muestra: Subconjunto representativo



## Tipos de muestreo:


1.   Aleatorio con reposici√≥n (Muerstra aleatoria simple **m.a.s**): Seleccionar una muestra de forma equiprobable (La equiprobabilidad es una situaci√≥n o evento en la teor√≠a de probabilidad, en que todos los resultados posibles son igualmente probables.‚Äã ) 



Screenshot_1.png

```python
from numpy import random as rd
numberElments = 10
numberSize = 8 
ary = list(range(numberElments))
```

```python
rd.choice(ary, size=numberSize, replace=True)
```

2.   Aleatorio sin reposici√≥n: No es equiprobable 


```python
rd.choice(ary, size=numberSize, replace=False)
```
